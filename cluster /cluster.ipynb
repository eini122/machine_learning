```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creat blobs

data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.6, random_state = 8)
transformation = [[0.6, -0.6], [-.4, 0.8]]
points= data[0]
print(data[1])
```

```python
plt.scatter(points[:,0], points[:,1], c=data[1], cmap='viridis')
plt.xlim(0,15)
plt.ylim(0,15)
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans
```

```python
kmeans.fit(points)
```

```python
print(kmeans.cluster_centers_)
```

```python
y_km = kmeans.fit_predict(points)
```

```python
plt.scatter(points[y_km==0,0], points[y_km==0,1], c='red')
plt.scatter(points[y_km==1,0], points[y_km==1,1], c='yellow')
plt.scatter(points[y_km==2,0], points[y_km==2,1], c='black')
plt.scatter(points[y_km==3,0], points[y_km==3,1], c='blue')
plt.xlim(-15,15)
plt.ylim(-15,15)
```
